```python
abstracts = df['Abstract'].tolist()

# Preprocessing
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    text = re.sub(r'\S*@\S*\s?', '', text)  # Remove emails
    text = re.sub(r'\s+', ' ', text)  # Remove newlines, extra whitespaces
    text = re.sub(r'\d+', '', text)  # Remove numbers
    text = text.lower()  # Convert to lowercase
    text = word_tokenize(text)  # Tokenize
    text = [word for word in text if word.isalpha()]  # Remove non-alphabetic tokens
    text = [word for word in text if word not in stop_words]  # Remove stopwords
    text = [lemmatizer.lemmatize(word) for word in text]  # Lemmatize
    return text

processed_abstracts = [preprocess_text(str(abstract)) for abstract in abstracts]

# Create dictionary and corpus for each abstract
topic_models = []
for i, abstract in enumerate(processed_abstracts):
    dictionary = corpora.Dictionary([abstract])
    corpus = [dictionary.doc2bow(abstract)]

    # Train LDA model
    lda_model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=1, passes=10)
    topic_models.append(lda_model)

# Print topics for each row
for i, model in enumerate(topic_models):
    print("Abstract", i+1)
    print("Topic:", model.print_topics()[0])
    print()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Load the dataframe containing the abstracts
df1 = pd.DataFrame({'Abstract': df['Abstract']})

# Remove rows with missing values
df1 = df1.dropna(subset=['Abstract'])

# Convert the 'Abstract' column to strings
df1['Abstract'] = df1['Abstract'].astype(str)

# Create a TfidfVectorizer to extract keywords from the abstracts
vectorizer = TfidfVectorizer(max_df=0.5, min_df=2, stop_words='english')
X = vectorizer.fit_transform(df1['Abstract'])

# Get the feature names (keywords)
keywords = vectorizer.get_feature_names_out()

# Create a list of keywords for each abstract
keyword_lists = []
for row in X.toarray():
    top_keywords = row.argsort()[-10:]  # Get the indices of the top 10 keywords
    keyword_list = '; '.join([keywords[i] for i in top_keywords])
    keyword_lists.append(keyword_list)

# Add the list of keywords to the dataframe
df1['Topic Modelling Keywords'] = keyword_lists

# Print the dataframe
print(df1)
```

```python
# Import libraries
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans


# Preprocess text
abstracts = df['Abstract'].str.lower().str.replace('[^\w\s]', '')
import re

#abstracts = abstracts.str.lower().apply(lambda x: re.sub(r'[^a-zA-Z\s]', '', x))
#abstracts = abstracts.str.lower().apply(lambda x: re.sub(r'[^a-zA-Z\s]', '', x))
abstracts = abstracts[abstracts.notnull() & abstracts.astype(str).notnull()]
abstracts = abstracts.map(lambda x: re.sub(r'[^a-zA-Z\s]', '', x.lower()))

# Create a TF-IDF vectorizer
vectorizer = TfidfVectorizer(stop_words='english')

# Fit the vectorizer to the abstracts
tfidf_matrix = vectorizer.fit_transform(abstracts)

# Cluster the abstracts using K-Means
num_clusters = 10
kmeans = KMeans(n_clusters=num_clusters)
kmeans.fit(tfidf_matrix)

# Extract titles from the clusters
titles = []
for i in range(num_clusters):
    cluster_abstracts = abstracts[kmeans.labels_ == i]
    most_frequent_words = vectorizer.get_feature_names_out()[np.argsort(kmeans.cluster_centers_[i])[:-10 - 1:-1]]
    titles.append(' '.join(most_frequent_words))

# Print the titles
print(titles)

# Create a dataframe
Topic_titles = pd.DataFrame(titles, columns=['Topic_Titles'])

# Print the dataframe
print(Topic_titles.to_string())
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Create a word cloud for each topic
for idx, title in enumerate(titles):
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(title)

    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(f'Topic {idx + 1}')
    plt.axis('off')
    plt.show()
```
